```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, date
```

```python
AEK201df = pd.read_pickle('../data/pickled_data/AEK201_all_data.pkl')
AFL259df = pd.read_pickle('../data/pickled_data/AFL259_all_data.pkl')
APK309df = pd.read_pickle('../data/pickled_data/APK309_all_data.pkl')
APK310df = pd.read_pickle('../data/pickled_data/APK310_all_data.pkl')

AEK201_well = pd.read_pickle('../data/pickled_data/AEK201_well_data.pkl')
AFL259_well = pd.read_pickle('../data/pickled_data/AFL259_well_data.pkl')
APK309_well = pd.read_pickle('../data/pickled_data/APK309_well_data.pkl')
APK310_well = pd.read_pickle('../data/pickled_data/APK310_well_data.pkl')
```

## AEK201

```python
## AEK209
df = AEK201df.copy()
df_well = AEK201_well.copy()

X = df.avg_well_depth.values
y = df.date.values

plt.figure(figsize=(12,6))
plt.title('Average Water Level', fontsize=16)
plt.plot(y,X, label = 'Water Level')
plt.ylabel('Depth from Land Surface in feet', fontsize=14)
plt.xlabel('Date', fontsize=14)
```

```python
# zooming in to look at continuity
low_date = datetime(2005,10,15)
discon_date = datetime(2006,2,15)

zoomeddf = df.loc[(df.date >= low_date) & (df.date <= discon_date)].copy()

## The target variable
X = zoomeddf.avg_well_depth.values
y = zoomeddf.date.values

plt.figure(figsize=(12,6))
plt.title('Average Water Level', fontsize=16)
plt.plot(y,X, label = 'Water Level')
plt.ylabel('Depth from Land Surface in feet', fontsize=14)
plt.xlabel('Date', fontsize=14)
```

```python
display(zoomeddf.tail(20))
display(df_well.tail(5))
```
